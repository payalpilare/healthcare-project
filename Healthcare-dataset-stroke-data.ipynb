# Data Visualization

# count plot showing Hypertension and Heart Disease analysis

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='hypertension', hue='heart_disease', data=df)
plt.title('Counts of Hypertension and Heart Disease')
plt.show()
```

# Box plot represents the BMI Distribution by Gender

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='bmi', data=df)
plt.title('BMI Distribution by Gender')
plt.show()
```

# count plot showing stroke count by Gender

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='stroke', hue='gender', data=df)
plt.title('Stroke Count by Gender')
plt.show()
```

# Heatmap showing the correlation with different aspects

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='viridis', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

# Pie chart represents the smoking status distribution

```python
plt.figure(figsize=(10, 6))
df['smoking_status'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Smoking Status Distribution')
plt.show()
```
